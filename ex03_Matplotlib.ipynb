```python
plt.plot(x, color='r', label='엑스')
plt.plot(y, color='k', label='와이')
plt.plot(z, color='m', label='지')
plt.xlabel('x축')
plt.ylabel('y축')
plt.title('x,y,z 그래프')
plt.legend()
plt.show()
```

### 차대차 사건 중 죽거나 다친 사람이 많은 발생지 시도를 알아보고 시각화

```python
# 1. 차대차 사건 중
# 2. (1) 죽거나 다친 사람(사상자)이 많은 (2) 발생지 시도
```

```python
data.head(2)
```

```python
acc=data[data['사고유형_대분류']=='차대차']
acc
```

```python
acc['발생지시도'].unique()
```

```python
acc['사상자수'].value_counts()
```

```python
acc.loc[:,['사상자수','발생지시도']]
```

```python
# 그룹화 = groupby('그룹화 시킬 컬럼')
# .sum() 이나  .mean() 가능
acc_g=acc[['사상자수','발생지시도']].groupby('발생지시도').sum()
acc_g
```

```python
x=acc_g.index
y=acc_g['사상자수']
```

```python
# 시각화
plt.figure(figsize=(10,5)) # 가로, 세로
plt.bar(x,y)

plt.show()
```

```python
plt.figure(figsize=(10,5)) # 가로, 세로
plt.bar(x,y)
plt.title('차vs차 교통사고의 사상자 수')
plt.xlabel('지역')
plt.ylabel('사상자 수')

plt.show()
```
